```python
import re
import logging
from slack_sdk import WebClient
```

```python
logger = logging.getLogger(__name__)
logging.getLogger().setLevel(logging.INFO)
```

```python
client = WebClient(token="")
```

```python
for result in client.conversations_list():
    for channel in result["channels"]:
        if channel['name'] == 'food':
            food_channel = channel
            channel_id = channel["id"]
            break
```

```python
history = client.conversations_history(channel=channel_id, limit=1000)
```

```python
restaurants = set({})
```

```python
for m in history['messages']:
    if "https://wolt.com" in m['text']:
        # wolt order
        if 'attachments' not in m:
            logging.warning(f"message without attachments: {m['text']}")
            continue
        
        link_text = m['attachments'][0]['fallback']
        restaurant_match = re.search('\| Order together from \s*([^\|]+) \|', link_text)
        if not restaurant_match:
            logger.error(f"regular expression didnt work on: {link_text}")
            continue
        
        restaurant_name = restaurant_match.groups()[0]
        logging.info(f"Found restaurant: {restaurant_name}")
        
```
